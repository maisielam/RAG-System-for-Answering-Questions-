```python
import arxiv
import os

def fetch_arxiv_papers(search_query, total_results=100, 
                       sort_by='relevance', sort_order='descending',
                       save_dir='/Users/tomorrowcute/DSCI521/MSDS_2024_2026/Side project/rag_langchain/data_source'):
    os.makedirs(save_dir, exist_ok=True)

    sort_criterion = arxiv.SortCriterion.Relevance if sort_by == 'relevance' else arxiv.SortCriterion.SubmittedDate

    print(f"\nFetching up to {total_results} papers for query: '{search_query}'")

    search = arxiv.Search(
        query=search_query,
        max_results=total_results,
        sort_by=sort_criterion,
        sort_order=arxiv.SortOrder.Descending if sort_order == 'descending' else arxiv.SortOrder.Ascending
    )

    downloaded = 0
    for result in search.results():
        paper_title = result.title.replace(' ', '_').replace('/', '_')
        file_path = os.path.join(save_dir, f"{paper_title}.pdf")

        if os.path.exists(file_path):
            print(f"Already downloaded: {paper_title}")
            continue

        print(f"Downloading: {result.title}")
        try:
            result.download_pdf(dirpath=save_dir, filename=f"{paper_title}.pdf")
            downloaded += 1
            print(f"Downloaded {downloaded}/{total_results} papers so far")
        except Exception as e:
            print(f"Failed to download {result.title}: {e}")

        if downloaded >= total_results:
            break

    print(f"\n Download complete. Total papers downloaded: {downloaded}")

if __name__ == "__main__":
    search_term = "machine learning"  # search term
    total_papers = 100  # Only 100 papers will be downloaded

    fetch_arxiv_papers(
        search_query=search_term,
        total_results=total_papers,
        sort_by='relevance',
        sort_order='descending'
    )
```
